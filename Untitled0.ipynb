```python
from google.colab import drive, files
import pandas as pd
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

# 1. Mount Google Drive
drive.mount('/content/drive')

# 2. Path Konfigurasi
base_path = '/content/drive/MyDrive/SKRIPSI/image_processing'
dataset_path = '/content/drive/MyDrive/SKRIPSI/image_processing/Dataset_Hadwriting.csv'

# Buat folder jika belum ada
image_folder = os.path.join(base_path, 'Image')
output_folder = os.path.join(base_path, 'Output')
model_folder = os.path.join(base_path, 'Models')
for f in [image_folder, output_folder, model_folder]:
    os.makedirs(f, exist_ok=True)

# 3. Load Dataset
df = pd.read_csv(dataset_path)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')] # Bersihkan kolom kosong

# Definisi Fitur untuk tiap Model (Sesuai Ketentuan Anda)
target_configs = {
    'Label_EI': ['Avg_Height', 'Word_Spacing'],
    'Label_SN': ['Baseline_Var', 'Margin_Var'],
    'Label_TF': ['Avg_Height', 'Stroke_Width'],
    'Label_JP': ['Baseline_Var', 'Word_Spacing']
}

print("Dataset Berhasil Dimuat dan Folder Siap!")
print(df.head())

print("--- LOG DATA LATIH ---")
print(f"Total Dataset: {len(df)} sampel")
print("Rata-rata fitur dataset (Mean):\n", df.mean(numeric_only=True))
```

```python
def extract_features_from_image(image_path):
    img = cv2.imread(image_path)
    if img is None: return None

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Binerisasi untuk memisahkan tulisan (hitam) dari background (putih)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Cari kontur (setiap huruf/kata)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    print(f"\n--- [LOG EKSTRAKSI CITRA: {os.path.basename(image_path)}] ---")

    # 1. Avg_Height
    heights = [cv2.boundingRect(c)[3] for c in contours if cv2.boundingRect(c)[3] > 5]
    avg_h = np.mean(heights) if heights else 0
    print(f"[1] Rata-rata Tinggi Huruf dari Avg_Height: {avg_h:.2f} piksel")

    # 2. Word_Spacing
    sorted_x = sorted([cv2.boundingRect(c) for c in contours], key=lambda x: x[0])
    gaps = [sorted_x[i+1][0] - (sorted_x[i][0] + sorted_x[i][2]) for i in range(len(sorted_x)-1)]
    avg_gap = np.mean([g for g in gaps if g > 0]) if gaps else 0
    print(f"[2] Jarak Rata-rata Kata dari Word_Spacing: {avg_gap:.2f} piksel")

    # 3. Baseline_Var
    baselines = [cv2.boundingRect(c)[1] + cv2.boundingRect(c)[3] for c in contours]
    sd_base = np.std(baselines) if baselines else 0
    print(f"[3] Standar Deviasi Baseline (Y-axis) dari Baseline_Var: {sd_base:.2f}")

    # 4. Stroke_Width
    dist_trans = cv2.distanceTransform(thresh, cv2.DIST_L2, 5)
    stroke_w = np.mean(dist_trans[dist_trans > 0])
    print(f"[4] Ketebalan Garis (Stroke) dari Stroke_Width: {stroke_w:.2f} piksel")

    # 5. Margin_Var (Varians Koordinat X awal baris)
    x_starts = [cv2.boundingRect(c)[0] for c in contours]
    var_x = np.var(x_starts) if x_starts else 0
    print(f"[5] Varians Margin (X-start) dari Margin_Var: {var_x:.2f}")

    # Return dictionary agar mudah dipetakan di Cell 4
    return {
        'Avg_Height': avg_h,
        'Word_Spacing': avg_gap,
        'Baseline_Var': sd_base,
        'Stroke_Width': stroke_w,
        'Margin_Var': var_x
    }, thresh

print("Fungsi Ekstraksi Fitur Berhasil Diperbarui.")
```

```python
models = {}
scalers = {}

print("--- [LOG TRAINING SVM] ---")
for label, feat_cols in target_configs.items():
    X = df[feat_cols].values

    # Mapping label
    if label == 'Label_EI':
        y = df[label].map({'E': 1, 'I': -1}).values
    elif label == 'Label_SN':
        y = df[label].map({'S': 1, 'N': -1}).values
    elif label == 'Label_TF':
        y = df[label].map({'T': 1, 'F': -1}).values
    elif label == 'Label_JP':
        y = df[label].map({'J': 1, 'P': -1}).values
    else:
        y = df[label].values

    sc = StandardScaler()
    X_scaled = sc.fit_transform(X)

    model = SVC(kernel='linear', C=1.0)
    model.fit(X_scaled, y)

    models[label] = model
    scalers[label] = sc

    # --- PINDAHKAN KE SINI (MASUK DALAM INDENTASI FOR) ---
    print(f"\n" + "="*40)
    print(f" TARGET LABEL : {label} ")
    print(f"="*40)
    print(f"  > Fitur Digunakan       : {feat_cols}")
    print(f"  > Rata-rata Fitur (Mean): {sc.mean_}")
    print(f"  > Std Deviasi Fitur     : {sc.scale_}")
    print(f"  > Bobot SVM (w)         : {model.coef_[0]}")
    print(f"  > Bias SVM (b)          : {model.intercept_[0]}")
    print(f"  STATUS: Model {label} Berhasil Dilatih!")
    print("-" * 40)

print("\n>>> SEMUA PROSES TRAINING SELESAI <<<")
```

```python
# 1. Interface Upload
print("\nSilahkan upload gambar tulisan tangan:")
uploaded = files.upload()

for filename in uploaded.keys():
    # Foto disimpan ke base_path/Image (image_folder)
    path_img = os.path.join(image_folder, filename)
    with open(path_img, 'wb') as f:
        f.write(uploaded[filename])
    print(f"Foto berhasil disimpan di: {path_img}")

    # 2. Jalankan Ekstraksi (Print rata-rata akan muncul di sini)
    features_dict, thresh_img = extract_features_from_image(path_img)

    if features_dict:
        print(f"\n--- [LOG PERHITUNGAN MATEMATIS SVM] ---")
        final_report = []

        for label, feat_cols in target_configs.items():
            # Ambil nilai fitur mentah berdasarkan kolom yang dibutuhkan model
            raw_val = np.array([features_dict[c] for c in feat_cols]).reshape(1, -1)

            # Tahap 1: Scaling (Z-Score)
            scaled_val = scalers[label].transform(raw_val)

            # Tahap 2: Decision Function (f(x) = wx + b)
            decision_val = models[label].decision_function(scaled_val)[0]
            pred = models[label].predict(scaled_val)[0]

            # Tahap 3: Interpretasi Label
            if label == 'Label_EI': res = "Extrovert (E)" if pred == 1 else "Introvert (I)"
            elif label == 'Label_SN': res = "Sensing (S)" if pred == 1 else "Intuition (N)"
            elif label == 'Label_TF': res = "Thinking (T)" if pred == 1 else "Feeling (F)"
            else: res = "Judging (J)" if pred == 1 else "Perceiving (P)"

            log_msg = (f"[{label.replace('Label_','')}]\n"
                       f"  > Fitur Mentah : {raw_val[0]}\n"
                       f"  > Target Fitur : {target_configs[label]}\n" # Corrected line
                       f"  > Hasil Scaling: {scaled_val[0]}\n"
                       f"  > f(x) = w.x+b : {decision_val:.4f}\n"
                       f"  > PREDIKSI     : {res}\n")
            print(log_msg)
            final_report.append(log_msg)

        # 3. Simpan Hasil ke Output
        txt_name = f"Laporan_{filename.split('.')[0]}.txt"
        with open(os.path.join(output_folder, txt_name), 'w') as f:
            f.write(f"LAPORAN ANALISIS GRAFOLOGI: {filename}\n" + "="*40 + "\n")
            f.writelines(final_report)

        # 4. Tampilkan Visualisasi
        plt.figure(figsize=(10,4))
        plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(cv2.imread(path_img), cv2.COLOR_BGR2RGB)); plt.title("Original")
        plt.subplot(1,2,2); plt.imshow(thresh_img, cmap='gray'); plt.title("Hasil Binerisasi")
        plt.show()

        print(f"Analisis Selesai. Laporan tersimpan di: {output_folder}/{txt_name}")
```
